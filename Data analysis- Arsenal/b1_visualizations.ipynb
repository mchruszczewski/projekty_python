```python
import pandas as pd
import numpy as np
import datetime as d
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import re 
from a2_extra_useful_columns_functions import *
from a3_data_analysis_masters import *
from a4_data_analysis_all_teams import*
```

```python
data= best_def_table.iloc[0:-1]
plt.figure(figsize=(9,9))
sns.barplot(data=data, x= 'season', y= 'xGA/90', color='gray')
plt.axhline(y=np.mean(best_def_table['xGA/90']), color= 'black', ls='--')
plt.xticks(rotation=30, size=12)
plt.yticks(np.arange(0, 1.1, step=0.1), size= 12)
plt.xlabel("Seasons",size= 12)
plt.ylabel("xGA/90", size=12)
plt.title("xGA/90 of a teams with the best defence", size= 12)
plt.savefig(r"xGA90.png", dpi = 500)
plt.show()
```

```python
data= all_teams_def
plt.figure(figsize=(24,16))
sns.scatterplot(data=data, x= 'GA', y="Rk", hue="Seasons", s=200)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.yticks(np.arange(1, 21, step=1), size= 15)
plt.xticks(np.arange(20,90, step= 5), size=15)
plt.xlabel("Goals Allowed",size= 15)
plt.ylabel("Positions", size=15)
plt.grid(ls ='--')
plt.legend(loc=2, prop={'size': 15})
plt.title("Goals For of all teams vs. their positions in League", size= 15)
plt.savefig(r"GA90.png", dpi = 500)
plt.show()
```
